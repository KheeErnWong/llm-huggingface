```python
from massive import RESTClient
from dotenv import load_dotenv
import os
import pandas as pd
```

```python
load_dotenv()
api_key = os.getenv("MASSIVE_API_KEY")
```

```python
client = RESTClient(api_key, pagination=False)
```

```python
ticker = "AAPL"
aggs = []
for a in client.list_aggs(
    ticker=ticker,
    multiplier=1,
    timespan="day",
    from_="2025-01-01",
    to="2025-11-21",
    limit=5,
):
    aggs.append(a)
```

```python
print(aggs)
```

```python
df = pd.DataFrame([a.__dict__ for a in aggs])
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")
```

```python
df.head()
```

```python
tickers = []
for t in client.list_tickers(
	market="stocks",
	active="true",
	order="asc",
	sort="ticker",
	):
    tickers.append(t)
```

```python
print(tickers)
```

```python
df = pd.DataFrame([a.__dict__ for a in tickers])
df.head()
```
